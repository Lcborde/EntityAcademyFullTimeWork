### Data Recoding - Assess data types and change if appropriate
#### listing all unique answers then converting categorical values to integers for statitical analysis.

```python
Hotels2
```

```python
def res_room_type_recode (letter):
    
    if letter == "A":
        return 0
    if letter == "B":
        return 1
    if letter == "C":
        return 2
    if letter == "D":
        return 3
    if letter == "E":
        return 4
    if letter == "F":
        return 5
    if letter == "G":
        return 6
    if letter == "H":
        return 7
    if letter == "L":
        return 8
    if letter == "P":
        return 9
    
Hotels2["res_room_typeR"] = Hotels2["reserved_room_type"].apply(res_room_type_recode)
```

```python
Hotels2
```

#### assigned_room_type recode

```python
def assn_room_type_recode (letter):
    
    if letter == "A":
        return 0
    if letter == "B":
        return 1
    if letter == "C":
        return 2
    if letter == "D":
        return 3
    if letter == "E":
        return 4
    if letter == "F":
        return 5
    if letter == "G":
        return 6
    if letter == "H":
        return 7
    if letter == "K":
        return 8
    if letter == "L":
        return 9
    if letter == "P":
        return 10
    
Hotels2["assn_room_typeR"] = Hotels2["assigned_room_type"].apply(assn_room_type_recode)
Hotels2
```

#### customer_type recode

```python
def cust_type_recode (series):
    
    if series == "Contract":
        return 0
    if series == "Group":
        return 1
    if series == "Transient":
        return 2
    if series == "Transient-Party":
        return 3

Hotels2["cust_typeR"] = Hotels2["customer_type"].apply(cust_type_recode)
Hotels2
```

#### reservation_status recode

```python
def res_status_recode (series):
    
    if series == "Check-Out":
        return 0
    if series == "Canceled":
        return 1
    if series == "No-Show":
        return 2
    
Hotels2["res_statusR"] = Hotels2["reservation_status"].apply(res_status_recode)
Hotels2
```

### hotel

```python
def hotel_recode (series):
    
    if series == "City Hotel":
        return 0
    if series == "city HOTEL":
        return 0
    if series == "CITY hotel":
        return 0
    if series == "Resort Hotel":
        return 1
    if series == "RESORT hotel":
        return 1
    if series == "Resort HOTEL":
        return 1
    
Hotels2["hotelR"] = Hotels2["hotel"].apply(hotel_recode)
Hotels2
```

### arrival_date_month

```python
def arrival_date_month_recode (series):
    
    if series == "January":
        return 0
    if series == "JANUARY":
        return 0
    if series == "February":
        return 1
    if series == "FEBRUARY":
        return 1
    if series == "March":
        return 2
    if series == "MARCH":
        return 2
    if series == "April":
        return 3
    if series == "APRIL":
        return 3
    if series == "May":
        return 4
    if series == "MAY":
        return 4
    if series == "June":
        return 5
    if series == "JUNE":
        return 5
    if series == "July":
        return 6
    if series == "July":
        return 6
    if series == "August":
        return 7
    if series == "AUGUST":
        return 7
    if series == "September":
        return 8
    if series == "SEPTEMBER":
        return 8
    if series == "October":
        return 9
    if series == "OCTOBER":
        return 9
    if series == "November":
        return 10
    if series == "NOVEMBER":
        return 10
    if series == "December":
        return 11
    if series == "DECEMBER":
        return 11
    
Hotels2["arrival_date_monthR"] = Hotels2["arrival_date_month"].apply(arrival_date_month_recode)
Hotels2
```

### Meal

```python
def meal_recode (series):
    
    if series == "BB":
        return 0
    if series == "HB":
        return 1
    if series == "SC":
        return 2
    if series == "FB":
        return 3
    if series == "Undefined":
        return 4
    
Hotels2["mealR"] = Hotels2["meal"].apply(meal_recode)
Hotels2
```

### deposit_type

```python
def deposit_type_recode (series):
    
    if series == "No Deposit":
        return 0
    if series == "Non Refund":
        return 1
    if series == "Refundable":
        return 2
    
Hotels2["deposit_typeR"] = Hotels2["deposit_type"].apply(deposit_type_recode)
Hotels2
```

### market_segment

```python
def market_segment_recode (series):
    
    if series == "Online TA":
        return 0
    if series == "Offline TA/TO":
        return 1
    if series == "Groups":
        return 2
    if series == "Direct":
        return 3
    if series == "Corporate":
        return 4
    if series == "Complementary":
        return 5
    if series == "Aviation":
        return 6
    if series == "Undefined":
        return 7
    
Hotels2["market_segmentR"] = Hotels2["market_segment"].apply(market_segment_recode)
Hotels2
```

### distribution_channel

```python
def dist_channel_recode (series):
    
    if series == "TA/TO":
        return 0
    if series == "Direct":
        return 1
    if series == "Corporate":
        return 2
    if series == "GDS":
        return 3
    if series == "Undefined":
        return 4

Hotels2["dist_channelR"] = Hotels2["distribution_channel"].apply(dist_channel_recode)
Hotels2
```

#### Recode into 3 categories GBR, PRT and then everything else.

```python
def country_recode (series):
    
    if series == "GBR":
        return 0
    if series == "PRT":
        return 1
    else:
        return 2
    
Hotels2["countryR"] = Hotels2["country"].apply(country_recode)
Hotels2
```
