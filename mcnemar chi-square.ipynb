### import packages

```python
import pandas as pd
import statsmodels as sm
from statsmodels.stats.contingency_tables import mcnemar
```

### load data

```python
pd.set_option('display.max_columns', None)
```

```python
hotels = pd.read_csv('/Users/chercher/Desktop/Entitiy apprenticeship /project 1 hotels/Data/final data/usethis.csv')
hotels
```

### set up the question -  Does reservation status and is canceled differ in any way?

#### data wrangling was done ahead of time to recode columns

### make a contingency table for res_statusR 0=checked out, 1= is canceled, 2= No show. For is_canceled 0=not canceled, 1=cancelled

```python
hotels_crosstab = pd.crosstab(hotels['res_statusR'], hotels['is_canceled'])
hotels_crosstab
```

### test assumptions = run analysis
##### In Python, there is no way to test the assumption of at least five expected per cell, which means that if you are analyzing high stakes data, where accuracy really matters, Python is NOT the program for you to run a McNemar Chi-Square in.

```python
result = sm.stats.contingency_tables.mcnemar(hotels_crosstab, exact=False, correction=True)
print(result)
```

### The results are significant - the p value is less than .05, so reservation status and is canceled are highly corrolated. I suspect the results of this test are not good. Instead of a number the statistic is "inf". reservation staus is the question and is_cancelled is a possible answer. this is why they are highly corrolated.

### How do they differ? There is no way to look further into this corrolation because Python does not offer the ability to look at standardized residuals, you can't look at post hocs.
